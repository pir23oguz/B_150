## Coklu data ve legend verme

```python
age = np.arange(25, 44, 2.5)
salary = np.arange(40000, 76000, 5000)

salary2 = [45372, 46852, 69401, 56790, 73000, 75000, 80000, 83454]
```

```python
plt.plot(age,salary, "b:", label="Salary")
plt.plot(age,salary2, "r-", label="Salary 2")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend()
plt.title("SALARY BY AGE");  # bunla calismadi
```

```python
plt.plot(age,salary, "b:", label="Salary")
plt.plot(age,salary2, "r-", label="Salary 2")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend()
plt.title("SALARY BY AGE");
```
