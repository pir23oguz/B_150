Hopkins testi, veri kmesinin dzgn (uniform) bir dalma sahip olup olmadn tahmin etmek iin kullanlan bir istatistiksel testtir.

Hopkins skoru 0'a yaklayorsa, bu, veri kmesinin dzgn bir dalma sahip olmad anlamna gelir. Bu durumda, veriyi snflandrmak iin kmeleme (clustering) yntemleri kullanmak yararl olabilir.

Ancak Hopkins skoru belirli bir deerin (rnein 0.3) zerindeyse, bu, veri kmesinin dzgn bir dalma sahip olduu anlamna gelir. Bu tr bir durumda, kmeleme yntemlerini kullanmak pek anlaml olmayabilir.

zetle, Hopkins testi, bir veri kmesinin dzgn bir dalm gsterip gstermediini ve bu balamda kmelemenin uygun olup olmadn deerlendirmenizi salar.

# Choosing The Optimal Number of Clusters

## Elbow metod

**Inertia (Kmeler i Toplam Kareler):** Inertia, her veri noktasnn kme merkezine olan uzaklnn karesinin toplamdr. Dk bir inertia deeri, veri noktalarnn kendi kme merkezlerine daha yakn olduunu gsterir, bu da genellikle iyi bir kmeleme olduunu dndrebilir. Ancak, inertia deeri, kmelerin says arttka genellikle azalr. Bu nedenle, farkl kme saylar iin inertia deerlerini karlatrarak en uygun kme saysn semek zordur.

```python
ssd = []

K = range(2,10)

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(X)
    ssd.append(model.inertia_)
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method") 
plt.show()
```

```python
ssd # sum of squared distance
```

```python
pd.Series(ssd).diff() # sadece DataFrame serieste alr. index srasna gre satrlarn farkn alr
```

```python
df_diff =pd.DataFrame(-pd.Series(ssd).diff()).rename(index = lambda x : x+1)
df_diff
```

## Silhouette analysis

Silhouette analizi, kmeleme performansn deerlendirmek iin kullanlan bir metriktir. Kme ii benzerlii ve kmeler aras benzerlii karlatrarak, kmelerin ne kadar iyi ayrldn gsterir. Silhouette deeri, her veri noktas iin hesaplanr ve bu deerlerin ortalamas alnarak genel bir kmeleme kalitesi lm elde edilir.

Silhouette deeri -1 ile 1 arasnda deiir:

1: i veri noktas kendi kmseinde iyi yerletirilmi ve dier kmelerden uzakta demektir.

0: i veri noktas iki kmse arasnda bir snra yakn demektir.

1: i veri noktas yanl bir kmseye atanm demektir.

bu deerlerin "eik" deerleri kesin bir ekilde belirlenmemitir. Ancak, pratikte, ortalama silhouette deerinin yksek olmas (genellikle 0.5'ten byk) kmelenmenin iyi olduunu gsterirken, dk deerler (genellikle 0.25'ten dk) kmelenmenin iyi olmadn veya daha iyi yaplabileceini gsterebilir.

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X, K_means_model.labels_)
```

```python
range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

Siluet Skoru: Grselde her bir veri noktas iin bir siluet deeri (katsay) vardr. Bu deerler -1 ile 1 arasnda deiir. 1'e yakn deerler iyi bir kmelemeyi, 0'a yakn deerler bir kmse snrna yakn olan veri noktalarn ve -1'e yakn deerler ise yanl bir kmseye atanm veri noktalarn gsterir.

Kmelerin Boyutu: Grselde her kme iin bir siluet grafii vardr. Bu grafikler, kmelerin ne kadar homojen olduunu ve kmeler arasndaki ayrmn ne kadar belirgin olduunu gsterir.

Krmz Noktal izgi: Bu izgi, tm veri noktalar iin ortalama siluet skorunu gsterir. Bu izginin zerindeki kmeler genellikle iyi bir performans gsterirken, bu izginin altndaki kmeler daha dk bir performans gsterir.

Kme Genilikleri: Her bir kme iin siluet genilii, o kmeye ait veri noktalarnn siluet katsaylarn gsterir. Eer tm kmeler benzer geniliklere sahipse, bu, kmelerin benzer younlukta ve boyutta olduunu gsterir. Farkl geniliklerde kmeler, kmelerin boyut ve younluklarnn farkl olduunu gsterir.

SilhouetteVisualizer' yorumlarken dikkat edilmesi gerekenler:

Geni siluetler (1'e yakn deerler) iyi bir kmeleme performansn gsterir.
Dar veya negatif deerlere sahip siluetler, bu veri noktalarnn muhtemelen yanl kmeye atandn gsterir.
Eer birok veri noktas ortalama siluet skorunun (krmz noktal izgi) altndaysa, bu kmse saysnn gzden geirilmesi gerektiini gsterebilir.
