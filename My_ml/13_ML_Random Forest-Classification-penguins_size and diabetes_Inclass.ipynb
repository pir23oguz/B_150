**Label Encoder**

Hiyerarsi olan yerde label encoder hatali sonuclar verebilir onlar icin ordinal encoder kullanacagiz.

**Ordinal Encoder**
Sirali degiskenler icin kullanilir. Biz kendimiz belirtiyoruz kucuklere 0 yaz, ortalara 1 yaz, buyuklere 2 yaz diyoruz.
Tree based modeller hiyerarsiye dayanikli oldugu icin, bunu yapabiliyoruz.

+ Regressyonda dummies.
+ Tree based encoding.
+ Ordinal encoding daha kapsaml. Her zaman kullanlabilir.
+ One hot encoding > dummies. (Dummies kullanacan yerde **one hot encoding** yap, machine learningde dummies yapma.)

* Tree base modellerde hiyeyrarsiden etkilenmedigi icin; hem ordinal hem label encoding kullanilabilirken regresyon, KNN ve SVM turu analizler hiyerarsiden etkilendigi icin ordinal encoding kullanilabilir.

+ **Feature importance tm encoderlar iyi sonu verir. Random forest bu encoding ile gl.**

## Train | Test Split

```python
X = df.drop('species',axis=1)
y = df['species']
```

```python
cat = X.select_dtypes("object").columns
cat
```

```python
X[cat]
```

```python
enc = OrdinalEncoder()
X[cat] = enc.fit_transform(X[cat])
X.head()

# genel kural btn kategorik datay numerie evirmemeiz gerekir. btn data daki kategorikleri numerie evirir.
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Modelling and Model Performance

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix

def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, recall_score, precision_score, f1_score
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.model_selection import GridSearchCV
```

```python
dt_model = DecisionTreeClassifier(random_state=101)
dt_model.fit(X_train, y_train)
eval_metric(dt_model, X_train, y_train, X_test, y_test)

# encoder yaptmzda sonular daha iyi kt  
```

```python
from sklearn.tree import plot_tree

def report_model(model):
    model_pred = model.predict(X_test)
    model_train_pred = model.predict(X_train)
    print('\n')
    print("Test Set")
    print(confusion_matrix(y_test, model_pred))
    print('\n')
    print(classification_report(y_test,model_pred))
    print('\n')
    print("Train Set")
    print(confusion_matrix(y_train, model_train_pred))
    print('\n')
    print(classification_report(y_train,model_train_pred))
    plt.figure(figsize=(12,8),dpi=100)
    plot_tree(model.estimators_[10], feature_names=X.columns, 
          class_names=df.species.unique(),
          filled = True,
          fontsize = 8);
```

```python
report_model(pruned_tree)
```

## Train | Test Split and Scalling

```python
X = df.drop("Outcome",axis=1)
y = df["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

## Modelling and Model Performance

### Decision Tree

```python
dt_model = DecisionTreeClassifier(class_weight = "balanced", random_state=101)
```

```python
dt_model.fit(X_train, y_train)
```

```python
eval_metric(dt_model, X_train, y_train, X_test, y_test) # overfit var ilk nce CV sonra GridSearch ile parametreleri vereceiz.
```

__Cross Validate__

```python
model = DecisionTreeClassifier(class_weight = "balanced", random_state=101)

scores = cross_validate(model, X_train, y_train, scoring = ['accuracy', 'precision', 'recall', 'f1'], cv = 10)
df_scores = pd.DataFrame(scores, index = range(1, 11))

df_scores.mean()[2:]
```

__GridSearch__

```python
param_grid = {"splitter":["best", "random"],
              "max_features":[None, 3, 5, 7],    #None= Butun feature' lari kullan
              "max_depth": [None, 4, 5, 6, 7, 8, 9, 10],
              "min_samples_leaf": [1, 2, 3, 5],
              "min_samples_split": [2, 3, 5, 7, 9, 15]}
```

```python
model = DecisionTreeClassifier(class_weight = "balanced", random_state=101)
dt_grid_model = GridSearchCV(estimator=model,
                            param_grid=param_grid,
                            scoring='recall',
                            n_jobs = -1, verbose = 2).fit(X_train, y_train)
```

```python
dt_grid_model.best_params_
```

```python
eval_metric(dt_grid_model, X_train, y_train, X_test, y_test)
```
