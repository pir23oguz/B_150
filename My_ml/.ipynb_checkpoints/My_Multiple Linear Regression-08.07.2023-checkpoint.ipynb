1.Fit: fit metodunu çağırdığınızda, model verilere uyarlanır veya veriler üzerinde eğitilir. Bu süreç, verilerin özelliklerini öğrenir ve bu özellikler daha sonra dönüştürme veya tahminler yapma gibi işlemler için kullanılır. Örneğin, StandardScaler'ın fit metodu, verinin ortalamasını ve standart sapmasını hesaplar.
2.Transform: transform metodu, fit metodu ile öğrenilen parametreleri kullanarak veriyi dönüştürür. Örneğin, StandardScaler'ın transform metodu, veriyi standardize eder yani her özelliği, özelliğin ortalamasını çıkarır ve standart sapmaya böler. Bu işlem, verinin ortalamanın 0 ve standart sapmanın 1 olduğu bir hale getirir.
3.Fit_transform: fit_transform metodu, fit ve transform metodlarını tek bir adımda gerçekleştirir. Yani önce fit metodu uygulanır ve model veriye uyarlanır, ardından aynı veri transform metodu ile dönüştürülür. Bu genellikle veri önişleme aşamasında kullanılır. İlk olarak verinin üzerinde fit_transform kullanılarak öğrenme ve dönüştürme yapılır. Ancak test setinde sadece transform metodu kullanılır çünkü test setinin özelliklerini eğitim setine göre dönüştürmek istersiniz.

# Adjusted R2 Score (Duzeltilmis R2 Skor)

```python
def adj_r2(y_test, y_pred, df):
    r2 = r2_score(y_test, y_pred)
    n = df.shape[0]   # number of observations gözlem
    p = df.shape[1]-1 # number of independent variables feature
    adj_r2 = 1 - (1-r2)*(n-1)/(n-p-1)
    return adj_r2
```

```python
adj_r2(y_test, y_pred, df)
```

```python
def adj_r2(y_test, y_pred, df):
    r2 = r2_score(y_test, y_pred)
    n = df.shape[0]   # number of observations gözlem
    p =100 # number of independent variables feature
    adj_r2 = 1 - (1-r2)*(n-1)/(n-p-1)
    return adj_r2
```

# Evaluation Model

```python
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    print("Model testing performance: ")
    print("---------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
```

```python
eval_metric(y_test, y_pred)
```

```python
residual = y_test - y_pred
```

```python
plt.figure(figsize = (10, 6))
sns.scatterplot(x= y_test, y = residual)
plt.axhline(y = 0, color = "r", linestyle = "--");
```

hata terimleri (residuallar) lineer regresyon modelinin uygunluğunu kontrol etmek için önemli bir özelliktir. Scatter plot üzerinde residualların dağılımını kontrol etmek, modelin varsayımlarının doğru olup olmadığını değerlendirmenin önemli bir yoludur. İşte residualların dağılımını değerlendirirken bakılması gereken bazı önemli noktalar:
1. Rastgele Dağılım: Residualların scatter plot üzerinde rastgele dağılması gerekmektedir. Bu durum, bağımsız değişkenler ve hata terimleri arasında herhangi bir ilişkinin olmadığını gösterir, yani modelin tüm bilgiyi yakaladığını ve hata terimlerinin tamamen rastgele olduğunu gösterir.
2. Sabit Varyans (Homoskedastisite): Residuallar, scatter plot üzerinde yatay bir çizgi etrafında eşit olarak dağılmalıdır. Bu, hata terimlerinin sabit bir varyansa sahip olduğunu, yani homoskedastik olduğunu gösterir. Eğer residuallar belirli bir modelin etrafında bir 'fan' şekli oluşturuyorsa, bu genellikle heteroskedastisite işaret eder ve bu durum modelin uygun olmadığını gösterir.
3. Normal Dağılım: Residualların normal bir dağılıma sahip olması beklenir. Bu, genellikle bir histogram veya bir QQ-plot ile kontrol edilir. Normal dağılımdan sapmalar, modelin verileri tam olarak yakalamadığını veya bazı aykırı değerlerin bulunduğunu gösterebilir.
Bu kontroller, modelin doğruluğunu ve güvenilirliğini kontrol etmek için önemlidir. Herhangi bir ihlal durumunda, veri önişleme tekniklerini gözden geçirmek, daha karmaşık bir model kullanmayı düşünmek veya belki de aykırı değerleri kontrol etmek gerekebilir.
